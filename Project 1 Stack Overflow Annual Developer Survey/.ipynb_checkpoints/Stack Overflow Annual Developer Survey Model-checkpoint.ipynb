```python
def len_BuyNewTool(data):
    '''
    function to count each BuyNewTool value 
    
    INPUT:
    data - origin dataframe
    
    
    OUTPUT:
    mydict - dict with "value":count
    '''
    
    possible_vals = ['Ask developers I know/work with', 'Other (please specify):', 
                 'Read ratings or reviews on third party sites like G2Crowd',
                 'Research companies that have advertised on sites I visit',
                'Research companies that have emailed me',
                 'Start a free trial',
                 'Visit developer communities like Stack Overflow'
                ]
    
    mydict = dict.fromkeys(possible_vals, 0)
    for row in range(len(data)):
        
        list_val = part3["BuyNewTool"][row].split(";")
    
        if 'Ask developers I know/work with' in list_val :
            mydict["Ask developers I know/work with"] += 1 
        elif 'Other (please specify):' in list_val : 
            mydict['Other (please specify):'] += 1 
            
        elif 'Read ratings or reviews on third party sites like G2Crowd' in list_val : 
            mydict['Read ratings or reviews on third party sites like G2Crowd'] += 1 
        
        elif 'Research companies that have advertised on sites I visit' in list_val : 
            mydict['Research companies that have advertised on sites I visit'] += 1 
            
        elif 'Research companies that have emailed me' in list_val : 
            mydict['Research companies that have emailed me'] += 1 
            
        elif 'Start a free trial' in list_val : 
            mydict['Start a free trial'] += 1 
        else:
            mydict['Visit developer communities like Stack Overflow'] += 1 
    
    return mydict
```

# 5- handel Missing Value 

## 5.1- Change type of Catogical to Float

```python
def handle_numric(df,column):
    '''
    some column we want it to be float insted of object
    
    INPUT:
    df - origin dataframe 
    column - column you want to change the datatype of it 
    
    
    OUTPUT:
    df - a new dataframe with change datatype of one column
    '''
    
    # fill the missing value by 0 insted of nan
    df[column] = df[column].fillna("0")

    # change string to number to make redy for change datatype
    df.loc[df[column].str.contains("Less than 1 year"),column] = "0.5"
    df.loc[df[column].str.contains("More than 50 years"),column] = "51"

    # change datatype
    df[column] = df[column].astype(float)

    
    return df
    
    
```

```python
final_data = handle_numric(final_data,"YearsCode")
final_data = handle_numric(final_data,"YearsCodePro")
```

## 5.2- handle missing value 

```python
def fill_nan(df):
    '''
    fill the nan value with mean and mode 
    
    INPUT:
    df - origin dataframe 
    
    OUTPUT:
    - new dataframe with no missing value  
    '''
    # select only categorical data and fill the missing value with mode
    categorical_data = df.select_dtypes(include=['object'])
    fill_mode = lambda col: col.fillna(col.mode()[0])
    categorical_data = categorical_data.apply(fill_mode, axis=0)
    
    # select only numeric data and fill the missing value with mean
    numric_data = df.select_dtypes(float)
    fill_mean = lambda col: col.fillna(col.mean())
    numric_data = numric_data.apply(fill_mean, axis=0)
    
    
    return pd.concat([categorical_data,numric_data],axis=1)
    
```

```python
Final_data = fill_nan(final_data)
```

```python
X = Final_data.drop("MainBranch",axis=1)
y = Final_data["MainBranch"]
```

# 6- Feature Encoding 
we will use Label Encoding for any categorical variables with only 2 categories and target column, then One-Hot Encoding for any categorical variables with more than 2 categories. to make number of feature small.

### 6.1 Lable Encoding


```python
# Create a label encoder object
le = LabelEncoder()
le_count = 0

# Iterate through the columns
for col in X:
    if X[col].dtype == 'object':
        # If 2 or fewer unique categories
        if len(list(X[col].unique())) <= 2:
            # Train on the training Small_data
            le.fit(X[col])
            # Transform both training and testing Small_data
            X[col] = le.transform(X[col])
            
            # Keep track of how many columns were label encoded
            le_count += 1
            
print('%d columns were label encoded.' % le_count)
```

```python
#Lable Encoding of Y 

# Create a label encoder object
le = LabelEncoder()
le.fit(y) 
y = le.transform(y)
```
